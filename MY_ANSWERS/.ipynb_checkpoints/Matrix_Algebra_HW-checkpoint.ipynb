1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.
   

```python
def matrixAddition(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.") # used raise for exception handling
    
    # result matrix is an empty matrix with the same dimensions as the passed matrices
    rows, cols = len(matrix1), len(matrix1[0])
    result = [[0 for _ in range(cols)] for _ in range(rows)]
    

    for i in range(rows):
        for j in range(cols):
            result[i][j] = matrix1[i][j] + matrix2[i][j] # matrix addition
    
    return result

# sample application of function
matrix1 = [[12, 2, 7], 
           [4, 17, 1]]
matrix2 = [[5, 8, 0], 
           [11, 50, 6]]

result = matrixAddition(matrix1, matrix2)
for row in result:
        print(row)
```

2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrixMultiplication(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Number of columns in the first matrix must be equal to the number of rows in the second matrix.") # used raise for exception handling

    # Determine the dimensions of the result matrix
    rows1, cols1 = len(matrix1), len(matrix1[0])
    rows2, cols2 = len(matrix2), len(matrix2[0])
    
    # Initialize an empty result matrix
    result = [[0 for _ in range(cols2)] for _ in range(rows1)]

    # Perform matrix multiplication
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

# sample application of function
matrix1 = [[12, 2], 
           [4, 17]]
matrix2 = [[5, 8], 
           [11, 6]]

result = matrixMultiplication(matrix1, matrix2)
for row in result:
        print(row)
```

3. The rule of distributivity states that given two matrices ùê¥ and ùêµ and a scalar, ùëò, then ùëò(ùê¥+ùêµ)=ùëòùê¥+ùëòùêµ. Instead of writing a proof mathematically, develop two codes for ùëò(ùê¥+ùêµ) and ùëòùê¥+ùëòùêµ.

```python
def scalarMultMatrixAdd(scalar, matrix_A, matrix_B):
    result = []
    for i in range(len(matrix_A)):
        row = []
        for j in range(len(matrix_A[0])):
            row.append(scalar * (matrix_A[i][j] + matrix_B[i][j]))
        result.append(row)
    return result

# function for kA + kB
def scalarMultMatricesThenAdd(scalar, matrix_A, matrix_B):
    result_A = []
    result_B = []

    for i in range(len(matrix_A)):
        row_A = []
        row_B = []
        for j in range(len(matrix_A[0])):
            row_A.append(scalar * matrix_A[i][j])
            row_B.append(scalar * matrix_B[i][j])
        result_A.append(row_A)
        result_B.append(row_B)

    return result_A, result_B

#sample application for both codes
matrix_A = [[0, 10], 
            [4, 4]]
matrix_B = [[12, 6], 
            [2, 8]]
k = 2

result = scalarMultMatrixAdd(k, matrix_A, matrix_B)
print("k(A + B):")
for row in result:
    print(row)

result = scalarMultMatricesThenAdd(k, matrix_A, matrix_B)
print("\nkA + kB:")
for row in result:
    print(row)
```

5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    # Check if the input is a valid matrix (2D list)
    if not all(isinstance(row, list) for row in matrix) or not all(len(row) == len(matrix) for row in matrix):
        raise ValueError("Input is not a valid square matrix.")

    # Calculate the trace (sum of diagonal elements)
    trace_sum = sum(matrix[i][i] for i in range(len(matrix)))

    return trace_sum

# Example usage:
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

result = matrix_trace(matrix)
print(result)
```
